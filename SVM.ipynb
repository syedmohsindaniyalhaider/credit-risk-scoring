### Model Training

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, median_absolute_error
from sklearn.model_selection import GridSearchCV
from tqdm import tqdm
```

```python
# SVM without hyperparameters

X = standardized_df[['Profit Margin', 'Return on Total Assets', 'Credit limit', 'Likelihood of failure(%)', 'Number of Employees', 'Gearing', 'Net Current Assets']]
y = standardized_df['Credit score']

#Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=49)

#Initialize SVR model

from sklearn.svm import SVR

#Initialize SVR model
svm_regressor = SVR()

#Fit SVR model to the training data
svm_regressor.fit(X_train, y_train)

#Predict on the test set
y_pred_svm = svm_regressor.predict(X_test)

#Compute evaluation metrics
mse_svm = mean_squared_error(y_test, y_pred_svm)
print("Support Vector Machine Mean Squared Error:", mse_svm)

accuracySVM = r2_score(y_test, y_pred_svm)
print("R-squared (Accuracy):", accuracySVM)
```

```python
# SVM with hyperparameters

# Define a parameter grid to search over
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf']
}

# Create an SVR model object
svm_regressor = SVR()

# Wrap the SVR model with GridSearchCV
grid_search = GridSearchCV(svm_regressor, param_grid, cv=5, scoring='neg_mean_squared_error')  # Use negative MSE for minimization

# Use tqdm to display progress bar during the search
print("Tuning hyperparameters...")
grid_search.fit(X_train, y_train)

# Get the best model and best parameters
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_

# Use the best model for prediction
y_pred_svm = best_model.predict(X_test)

# Evaluate the model's performance with the best parameters
mse_svm = mean_squared_error(y_test, y_pred_svm)
accuracySVM = r2_score(y_test, y_pred_svm)
print("Support Vector Machine Mean Squared Error:", mse_svm)
print("R-squared (Accuracy):", accuracySVM)
```

```python
# Features that have the most significant impact on the prediction model after training/testing

svm_regressor = SVR(kernel='linear')
svm_regressor.fit(X_train, y_train)

#Extract coefficients
coefficients = svm_regressor.coef_.flatten()

#Pair features with their coefficients
feature_coefficient_list = list(zip(X.columns, coefficients))

#Sort feature coefficients in descending order
feature_coefficient_list.sort(key=lambda x: abs(x[1]), reverse=True)

#Print feature coefficients
for feature, coefficient in feature_coefficient_list:
    print(f"{feature}: {coefficient}")
```

```python
#Plot feature importances
features = [x[0] for x in feature_coefficient_list]
coefficients = [abs(x[1]) for x in feature_coefficient_list]

plt.figure(figsize=(12, 8))
plt.barh(features, coefficients, color='skyblue')
plt.xlabel('Coefficient')
plt.ylabel('Features')
plt.title('Feature Importances (SVM)')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Calculate Mean Squared Error (MSE)
mse_svm = mean_squared_error(y_test, y_pred_svm)

# Calculate Root Mean Squared Error (RMSE)
rmse_svm = np.sqrt(mse_svm)

# Calculate R-squared (Coefficient of Determination)
r_squared_svm = r2_score(y_test, y_pred_svm)


# Calculate Mean Absolute Error (MAE)
mae_svm = mean_absolute_error(y_test, y_pred_svm)


evaluation_df = pd.DataFrame({
    'Model': ['SVM'],
    'Mean Squared Error (MSE)': [mse_svm],
    'Root Mean Squared Error (RMSE)': [rmse_svm],
    'R-squared (Accuracy)': [r_squared_svm],
    'Mean Absolute Error (MAE)': [mae_svm],
})

evaluation_df
```
